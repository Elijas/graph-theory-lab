# Užduotis 4

Žinome, kai turime lyginį skaičių viršūnių su nelyginiu laipsniu, tuomet galime sudaryti baigtinį paprastąjį grafą (pgl. <i>Handshaking lemma</i>). Duotame grafe $2,2,3,3,4$ būtent taip ir yra (turime dvi viršūnes su laipsniu $3$).

Todėl toks penktos eilės grafas <b>egzistuoja</b>.

```r
g <- graph_from_literal(
    a-b, a-c, a-d, a-e,
    b-c, b-d,
    c-e
)
plot(g)
```
